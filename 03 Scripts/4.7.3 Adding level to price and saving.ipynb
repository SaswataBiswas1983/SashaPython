### First compress the required dtypes
#### Department_id to int 8 and aisle id to int 16

```python
df_new['department_id']= df_new['department_id'].astype('int8')
```

```python
df_new['aisle_id']= df_new['aisle_id'].astype('int16')
```

## Using the LOC function to apply the range of products

```python
df_new.loc[df_new['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_new.loc[(df_new['prices'] <= 15) & (df_new['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_new.loc[df_new['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```
